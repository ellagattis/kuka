# nearest neighbour search

```
#use scipy's cdist function to find closest magnaprobe point for one kuka point
def closest_point(point, magna_points):
    magna_locations = (magna_points.xc, magna_points.yc)
    magna_locations = (np.transpose(magna_locations))
    closest_index = (distance.cdist([point], magna_locations)).argmin()
    #result = np.count_nonzero(dist_matrix<=r)
    return magna_points.iloc[closest_index]
```

```
#ku nearest neighbours
ku_closest_points_list = pd.DataFrame()
ku_loc = (test.ku_x, test.ku_y)
ku_loc = np.transpose(ku_loc)

for i in range(len(ku_loc)):
  mag_cp = closest_point(ku_loc[i], mag)
  ku_closest_points_list = ku_closest_points_list.append(mag_cp)

#scale magnaprobe depths by 0.8 to account for velocity change in snow
ku_scaled_mag_depths = ku_closest_points_list.DepthCm*0.7
```

```
#ku shifts

shifted_echoes_ku_hv=[]
shifted_echoes_ku_vh=[]
shifted_echoes_ku_hh=[]
shifted_echoes_ku_vv=[]

for i in range(0,ku_vv_echo.shape[1],1):
    try: 
        #print(i)                   creating copies of echoes
        a=ku_vv_echo[:,i].copy()
        b=ku_hh_echo[:,i].copy()
        c=ku_hv_echo[:,i].copy()
        d=ku_vh_echo[:,i].copy()
        # creating mask m to find maxima above threshold -80
        m=np.r_[True, a[1:] > a[:-1]] & np.r_[a[:-1] > a[1:], True] & np.r_[a[:]>-80]
        indexes = [index for index in range(len(a)) if m[index] == True]
        shifted_echoes_ku_vv.append(np.roll(a,100-indexes[0]))
        shifted_echoes_ku_hh.append(np.roll(b,100-indexes[0]))
        shifted_echoes_ku_hv.append(np.roll(c,100-indexes[0]))
        shifted_echoes_ku_vh.append(np.roll(d,100-indexes[0]))

        #plt.plot(np.roll(a,100-indexes[0]),alpha=0.2)
    except:
        #print('issue',i)
        shifted_echoes_ku_vv.append(np.zeros((262)))
        shifted_echoes_ku_hh.append(np.zeros((262)))
        shifted_echoes_ku_hv.append(np.zeros((262)))       
        shifted_echoes_ku_vh.append(np.zeros((262))) 
```

```
#calculate Ku mask 

temp=shifted_echoes_ku_vv_array
temp1=np.roll(temp,shift=1,axis=0)
diff=temp1-temp
std=[]
for i in range (shifted_echoes_ku_vv_array.shape[0]):
    std.append(np.sum(diff[i,:]**2))
 
mask_ku=np.outer(np.asarray(std),np.ones((262)))
mask_ku.shape
mask_ku[mask_ku<10000]=np.nan
mask_ku[mask_ku>=10000]=1.
# mask_ku[:,1300]

#calculate Ka mask

temp=shifted_echoes_ka_vv_array
temp1=np.roll(temp,shift=1,axis=0)
diff=temp1-temp
std=[]
for i in range (shifted_echoes_ka_vv_array.shape[0]):
    std.append(np.sum(diff[i,:]**2))

mask_ka=np.outer(np.asarray(std),np.ones((437)))

mask_ka[mask_ka<20000]=np.nan
mask_ka[mask_ka>=20000]=1.
# mask[:,1300]    
```

```
# https://github.com/kapil-varshney/utilities/blob/master/training_plot/training_plot_ex_with_cifar10.ipynb
class TrainingPlot(keras.callbacks.Callback):
   
    # This function is called when the training begins
    def on_train_begin(self, logs={}):
        # Initialize the lists for holding the logs, losses and accuracies
        self.losses = []
        self.val_losses = []
        self.logs = []
    
    # This function is called at the end of each epoch
    def on_epoch_end(self, epoch, logs={}):
        
        # Append the logs, losses and accuracies to the lists
        self.logs.append(logs)
        self.losses.append(logs.get('loss'))
        self.val_losses.append(logs.get('val_loss'))
        
        # Before plotting ensure at least 2 epochs have passed
        if len(self.losses) > 1:
            
            # Clear the previous plot
            clear_output(wait=True)
            N = np.arange(0, len(self.losses))
            
            # You can chose the style of your preference
            # print(plt.style.available) to see the available options
            plt.style.use("seaborn")
            
            # Plot train loss, train acc, val loss and val acc against epochs passed
            plt.figure(figsize=(15,10))
            plt.plot(N, self.losses, label = "train_loss")
            plt.plot(N, self.val_losses, label = "val_loss")
            plt.title("Training Loss [Epoch {}]".format(epoch), size = 35)
            plt.xlabel("Epoch #", size = 30)
            plt.ylabel("Loss", size = 30)
            plt.legend(prop={'size':20})
            plt.show()

plot_losses = TrainingPlot()
```
